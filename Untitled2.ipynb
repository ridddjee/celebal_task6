Q1: Explain the roles of the Driver, Cluster Manager, and Executor in a Spark application.

The Driver is the main part of a Spark application that creates jobs, schedules tasks, and collects the final results. The Cluster Manager manages the available resources and assigns them to the application. The Executors run the tasks on worker machines, process the data, and send the results back to the Driver.

Q2: How does Spark’s Lazy Evaluation strategy improve performance when chain-processing large datasets?

Spark uses Lazy Evaluation, which means it does not execute transformations immediately. Instead, it records all the operations and runs them only when an action like show() or count() is called. This helps Spark optimize the execution plan using a DAG, reducing unnecessary work and improving performance on large datasets.

Q11: What is the difference between Transformations and Actions? Provide two examples of each.

Transformations are operations like filter() and select() that create a new DataFrame but are not executed immediately. Actions, such as show() and count(), trigger the execution of all pending transformations and produce the final result.

Q12: Write the Spark command to load a Parquet file from "path/to/input", filter out any rows where user_id is null, and save the result as a CSV at "path/to/output".

```python
df = spark.read.parquet("path/to/input")

df.filter(df.user_id.isNotNull()) \
  .write \
  .mode("overwrite") \
  .option("header", "true") \
  .csv("path/to/output")
```

Q13: In Spark Architecture, what is the difference between Client Mode and Cluster Mode?

In Client Mode, the Driver program runs on the user's local machine, making it suitable for development and testing. In Cluster Mode, the Driver runs inside the cluster, allowing the application to continue even if the user's machine disconnects, which makes it better for production environments.

Q14: Write a query to filter a dataset for rows where the region is 'North' OR the priority is 'High'.

Q15: When exploring a dataset, why is it safer to use .show(5) instead of .collect() on a multi-terabyte dataset?

Using .show(5) is safer because it displays only the first five rows without loading the entire dataset into the Driver's memory. On very large datasets, .collect() retrieves all the data to the Driver, which can consume a lot of memory and may even cause the application to fail.
